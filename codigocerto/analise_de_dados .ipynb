```python
# Importar bibliotecas 
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Carregamento e Exploração de Dados:
df = pd.read_csv(r"C:\Users\herme\OneDrive\Área de Trabalho\codigocerto\dados.csv", delimiter=',')
# Exibir as primeiras linhas
df.head()
```

```python
# Estatísticas descritiva
df.describe()
```

```python
plt.figure(figsize=(12, 6))
sns.countplot(y='Nome do Curso', data=df)
plt.show()
```

```python

sns.set(style="whitegrid")
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Quantidade de Vendas', y='Preço Unitário', data=df, s=100, alpha=0.6, edgecolor='w')
plt.title('Relação entre Quantidade de Vendas e Preço Unitário', fontsize=16, fontweight='bold')
plt.xlabel('Quantidade de Vendas', fontsize=14)
plt.ylabel('Preço Unitário', fontsize=14)
plt.show()
```

```python
# Calcular a receita total
df['Receita']=df['Quantidade de Vendas']*df['Preço Unitário']
receitaTotal = df['Receita'].sum()
print(f'Receita Total: R$ {receitaTotal}')
```

```python
# Identificar o curso com o maior número de vendas
curso_mais_vendido = df.groupby('Nome do Curso')['Quantidade de Vendas'].sum().idxmax()
print(f'Curso com maior número de vendas: {curso_mais_vendido}')
```

```python
# Visualizar a distribuição das vendas ao longo do tempo através de gráficos
data = {
    'Data': ['2023-01-01', '2023-01-02', '2023-01-03', '2023-01-04', '2023-01-05'],
    'Quantidade de Vendas': [20, 35, 30, 35, 27]
}
df = pd.DataFrame(data)
df['Data'] = pd.to_datetime(df['Data'])
df.set_index('Data', inplace=True)
sns.set(style="whitegrid")
plt.figure(figsize=(10, 6))
sns.barplot(x=df.index, y=df['Quantidade de Vendas'], hue=df.index, palette="Blues_d", dodge=False, legend=False)
plt.title('Distribuição das Vendas ao Longo do Tempo', fontsize=16)
plt.xlabel('Data', fontsize=14)
plt.ylabel('Quantidade de Vendas', fontsize=14)
plt.xticks(rotation=45)
for index, value in enumerate(df['Quantidade de Vendas']):
    plt.text(index, value + 0.5, str(value), color='black', ha="center")
plt.show()
```
